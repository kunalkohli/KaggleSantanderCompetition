```python
#X_train, X_test, y_train, y_test 
from keras.models import Model
from keras.layers import *
from keras import metrics


from keras.callbacks import Callback
def jacek_auc(y_true, y_pred):
    score, up_opt = tf.metrics.auc(y_true, y_pred)
    #score, up_opt = tf.contrib.metrics.streaming_auc(y_pred, y_true)    
    K.get_session().run(tf.local_variables_initializer())
    with tf.control_dependencies([up_opt]):
        score = tf.identity(score)
    return score

model = Sequential()

train_labels = to_categorical(y_train)
test_labels = to_categorical(y_test)

model.add(Dense(3200,input_dim=X_train.shape[1],  kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
model.add(Dropout(0.5, noise_shape=None, seed=None))

model.add(Dense(2400,  kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
model.add(Dropout(0.5, noise_shape=None, seed=None))

model.add(Dense(1600,  kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
model.add(Dropout(0.4, noise_shape=None, seed=None))

model.add(Dense(800, kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
model.add(Dropout(0.4, noise_shape=None, seed=None))

model.add(Dense(400,  kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
model.add(Dropout(0.5, noise_shape=None, seed=None))

model.add(Dense(200,  kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
model.add(Dropout(0.5, noise_shape=None, seed=None))

model.add(Dense(100,  kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
model.add(Dropout(0.5, noise_shape=None, seed=None))


model.add(Dense(50, kernel_initializer = glorot_uniform(seed=42)))
model.add(BatchNormalization())
model.add(LeakyReLU(alpha=0.3))
# model.add(LeakyReLU(alpha=.001))
model.add(Dropout(0.3, noise_shape=None, seed=None))

model.add(Dense(2, activation='softmax'))

opti = optimizers.Adamax(beta_1=0.9, beta_2=0.999, epsilon=0.01, decay=0.0)

model.compile(loss='binary_crossentropy', optimizer=opti, metrics=['acc']) #[jacek_auc])

es = EarlyStopping(monitor='val_loss', mode='min', patience=15)

class_weight = {1: 10,
                0: 0.55529771}

history = model.fit(X_train, train_labels,validation_split=0.2,
           epochs=200, batch_size=10000,  shuffle=True)
          #,class_weight=class_weight)
    
```

```python
score = model.evaluate(X_test, test_labels, batch_size=256)
print(score)

from sklearn.metrics import average_precision_score
from sklearn.metrics import roc_curve, auc,recall_score,precision_score
predictions = model.predict(X_test)
score = average_precision_score(test_labels, predictions)
print('area under the precision-recall curve: {:.6f}'.format(score))

score = precision_score(y_test, predictions[:,1].round())
print('precision score: {:.6f}'.format(score))

score = recall_score(y_test, predictions[:,1].round())
print('recall score: {:.6f}'.format(score))


from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
probas_ = model.predict_proba(X_test)
fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])

plt.plot(fpr, tpr, lw=1)

plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

plt.show()

roc_auc_clf = auc(fpr, tpr)
print(roc_auc_clf)

from sklearn.metrics import precision_recall_fscore_support as score
from sklearn import metrics

precision, recall, fscore, support = score(y_test, predictions[:,1].round())

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))
print('support: {}'.format(support))  
conf_matrix = metrics.confusion_matrix(y_test, predictions[:,1].round())

print (conf_matrix)
```

```python
#LGB
param = {
    'bagging_freq': 5,          
    'bagging_fraction': 0.38,   'boost_from_average':'false',   
    'boost': 'gbdt',             'feature_fraction': 0.04,     'learning_rate': 0.0085,
    'max_depth': -1,             'metric':'auc',                'min_data_in_leaf': 80,     'min_sum_hessian_in_leaf': 10.0,
    'num_leaves': 13,            'num_threads': 8,              'tree_learner': 'serial',   'objective': 'binary',
    'reg_alpha': 0.1302650970728192, 'reg_lambda': 0.3603427518866501,'verbosity': 1
}
# X_train, X_test, y_train
#X_train_balanced
#y_train_balanced
folds = StratifiedKFold(n_splits=12, shuffle=False, random_state=99999)
oof = np.zeros(len(data_train))
predictions_lgb = np.zeros(len(test_data))
features = [c for c in data_train.columns if c not in ['ID_code', 'target']] #basic features
target = data_train['target']

for fold_, (trn_idx, val_idx) in enumerate(folds.split(data_train.values, target.values)):
    print("Fold {}".format(fold_))
    trn_data = lgb.Dataset(data_train.iloc[trn_idx][features], label=target.iloc[trn_idx])
    val_data = lgb.Dataset(data_train.iloc[val_idx][features], label=target.iloc[val_idx])
    clf = lgb.train(param, trn_data, 1000000, valid_sets = [trn_data, val_data], verbose_eval=5000, early_stopping_rounds = 2000)
    oof[val_idx] = clf.predict(data_train.iloc[val_idx][features], num_iteration=clf.best_iteration)
    predictions_lgb += clf.predict(test_data[features], num_iteration=clf.best_iteration) / folds.n_splits
```

```python

test_data['target'] = 0.3*model.predict(test_data.loc[:, test_data.columns != 'ID_code'])[:,1] + 0.7* predictions_lgb
print(test_data['target'].value_counts())
test_data[['ID_code','target']].to_csv('submission_keras3.csv',index=False)

test_data = test_data.drop('target',axis=1)
test_data['target'] = 0.7*model.predict(test_data.loc[:, test_data.columns != 'ID_code'])[:,1] + 0.3* predictions_lgb
print(test_data['target'].value_counts())
test_data[['ID_code','target']].to_csv('submission_keras4.csv',index=False)

test_data = test_data.drop('target',axis=1)
test_data['target'] = 0.5*model.predict(test_data.loc[:, test_data.columns != 'ID_code'])[:,1] + 0.5* predictions_lgb
print(test_data['target'].value_counts())
test_data[['ID_code','target']].to_csv('submission_keras_balanced.csv',index=False)

test_data = test_data.drop('target',axis=1)
test_data['target'] = model.predict(test_data.loc[:, test_data.columns != 'ID_code'])[:,1]
print(test_data['target'].value_counts())
test_data[['ID_code','target']].to_csv('submission_keras_only_nn.csv',index=False)
```
